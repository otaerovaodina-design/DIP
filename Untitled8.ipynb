```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image (replace with your image path)
image = cv2.imread("pencil.jpg", cv2.IMREAD_GRAYSCALE)

# Histogram Equalization
equalized = cv2.equalizeHist(image)

# Show images
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title("Original")

plt.subplot(1,2,2)
plt.imshow(equalized, cmap='gray')
plt.title("Equalized")
plt.show()

# Show histograms
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.hist(image.ravel(), 256, [0,256])
plt.title("Original Histogram")

plt.subplot(1,2,2)
plt.hist(equalized.ravel(), 256, [0,256])
plt.title("Equalized Histogram")
plt.show()
```

```python
box = np.ones((5,5), np.float32) / 25
box_filtered = cv2.filter2D(image, -1, box)

gaussian3 = cv2.GaussianBlur(image, (3,3), 0)
gaussian5 = cv2.GaussianBlur(image, (5,5), 0)

plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.imshow(image, cmap='gray')
plt.title("Original")

plt.subplot(2,2,2)
plt.imshow(box_filtered, cmap='gray')
plt.title("Box Filter 5x5")

plt.subplot(2,2,3)
plt.imshow(gaussian3, cmap='gray')
plt.title("Gaussian 3x3")

plt.subplot(2,2,4)
plt.imshow(gaussian5, cmap='gray')
plt.title("Gaussian 5x5")

plt.show()
```

```python
blurred = cv2.GaussianBlur(image, (9,9), 10)

unsharp = cv2.addWeighted(image, 1.5, blurred, -0.5, 0)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title("Original")

plt.subplot(1,2,2)
plt.imshow(unsharp, cmap='gray')
plt.title("Unsharp Masking")

plt.show()
```

```python
import copy

def box_filt(n):
    kernel = np.ones((n, n), np.float32) / (n * n)
    return kernel

def apply_filters(image_input, box, filt_size):
    pad_size = int(np.ceil(filt_size/2))
    image_padded = np.pad(image_input,
                          ((pad_size,pad_size),(pad_size,pad_size)),
                          mode='symmetric')
    image_box = copy.deepcopy(image_input)
    row, column = image_input.shape

    for i in range(row):
        for j in range(column):
            patch = image_padded[i:i+filt_size, j:j+filt_size]
            image_box[i,j] = np.sum(box * patch)

    return image_box

image_input = cv2.imread("pencil.jpg",0).astype('float32')/255

filt_size = 7
k = 2

box = box_filt(filt_size)
image_blur = apply_filters(image_input, box, filt_size)
image_diff = image_input - image_blur
image_sharp = image_input + k * image_diff

plt.imshow(image_sharp, cmap='gray')
plt.title("Sharpened Image")
plt.show()
```

Report
In this lab I applied histogram equalization to improve contrast, used box and Gaussian filters to smooth and reduce noise, and performed unsharp masking to enhance edges, observing that larger kernel sizes increase blurring while higher k values increase sharpness but may add noise.
