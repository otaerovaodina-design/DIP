```
import numpy as np
import cv2
import matplotlib.pyplot as plt
def show(img, title=""):
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
kernel_x = np.array([[-1, 0, 1]])

img_square_shifted = np.zeros((200, 200), dtype=np.uint8)
img_square_shifted[80:140, 100:160] = 255

grad_shifted = cv2.filter2D(img_square_shifted, cv2.CV_64F, kernel_x)
grad_shifted_abs = np.abs(grad_shifted).astype(np.uint8)

show(img_square_shifted, "Shifted square")
show(grad_shifted_abs, "Gradient of shifted square")
```

```
import numpy as np
import cv2
import matplotlib.pyplot as plt

def show(img, title=""):
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()

img_square = np.zeros((200, 200), dtype=np.uint8)
img_square[50:150, 50:150] = 255

img_circle = np.zeros((200, 200), dtype=np.uint8)
cv2.circle(img_circle, (100, 100), 40, 255, -1)

sobel_square_x = cv2.Sobel(img_square, cv2.CV_64F, 1, 0, ksize=3)
sobel_square_y = cv2.Sobel(img_square, cv2.CV_64F, 0, 1, ksize=3)
sobel_square_mag = np.sqrt(sobel_square_x**2 + sobel_square_y**2)

sobel_circle_x = cv2.Sobel(img_circle, cv2.CV_64F, 1, 0, ksize=3)
sobel_circle_y = cv2.Sobel(img_circle, cv2.CV_64F, 0, 1, ksize=3)
sobel_circle_mag = np.sqrt(sobel_circle_x**2 + sobel_circle_y**2)

show(img_square, "Square")
show(np.clip(sobel_square_mag, 0, 255).astype(np.uint8), "Sobel of square")

show(img_circle, "Circle")
show(np.clip(sobel_circle_mag, 0, 255).astype(np.uint8), "Sobel of circle")
```

```
import numpy as np
import cv2
import matplotlib.pyplot as plt

def show(img, title=""):
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()

img_step = np.zeros((200, 200), dtype=np.uint8)
img_step[:, 100:] = 255

lap_step = cv2.Laplacian(img_step, cv2.CV_64F)

sobel_x = cv2.Sobel(img_step, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(img_step, cv2.CV_64F, 0, 1, ksize=3)
sobel_mag = np.sqrt(sobel_x**2 + sobel_y**2)

show(np.abs(lap_step).astype(np.uint8), "Laplacian")
show(np.clip(sobel_mag, 0, 255).astype(np.uint8), "Sobel magnitude")
```

```
import numpy as np
import cv2
import matplotlib.pyplot as plt

def show(img, title=""):
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()

img_square = np.zeros((200, 200), dtype=np.uint8)
img_square[50:150, 50:150] = 255

edges_1 = cv2.Canny(img_square, 30, 90)
edges_2 = cv2.Canny(img_square, 50, 150)
edges_3 = cv2.Canny(img_square, 100, 200)
edges_4 = cv2.Canny(img_square, 150, 250)

show(edges_1, "Canny 30-90")
show(edges_2, "Canny 50-150")
show(edges_3, "Canny 100-200")
show(edges_4, "Canny 150-250")
```
